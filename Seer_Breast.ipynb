#   IQR ile aykiri veri hesaplama
   Alt sınır: Q1 - 1.5 * IQR  
   Üst sınır: Q3 + 1.5 * IQR  
   Veri noktaları bu sınırların dışındaysa, aykırı değer olarak kabul edilir.

```python
# Sayısal bir sütun için IQR hesaplama
Q1 = data['Age'].quantile(0.25)  # 1. Çeyrek
Q3 = data['Age'].quantile(0.75)  # 3. Çeyrek
IQR = Q3 - Q1  # IQR hesaplama

# Alt ve üst sınırları hesaplama
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Aykırı değerleri belirleme
outliers = data[(data['Age'] < lower_bound) | (data['Age'] > upper_bound)]

# Aykırı değerlerin sayısını yazdırma
print(f"'Age' icin aykırı Değerlerin Sayısı: {outliers.shape[0]}")
```

```python
# Sayısal bir sütun için IQR hesaplama
Q1 = data['Tumor Size'].quantile(0.25)  # 1. Çeyrek
Q3 = data['Tumor Size'].quantile(0.75)  # 3. Çeyrek
IQR = Q3 - Q1  # IQR hesaplama

# Alt ve üst sınırları hesaplama
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Aykırı değerleri belirleme
outliers = data[(data['Tumor Size'] < lower_bound) | (data['Tumor Size'] > upper_bound)]

# Aykırı değerlerin sayısını yazdırma
print(f"'Tumor Size' icin aykırı Değerlerin Sayısı: {outliers.shape[0]}")

```

```python
# Aykırı değerleri sınırlarla değiştirme
data['Tumor Size'] = np.where(
    data['Tumor Size'] > upper_bound, upper_bound,
    np.where(data['Tumor Size'] < lower_bound, lower_bound, data['Tumor Size'])
)

# Aykırı değerlerin tekrar kontrolü
outliers_cleaned = data[(data['Tumor Size'] < lower_bound) | (data['Tumor Size'] > upper_bound)]
print(f"Aykırı Değerlerin Sayısı (Temizlendikten Sonra): {outliers_cleaned.shape[0]}")
```

```python
# Sayısal bir sütun için IQR hesaplama
Q1 = data['Regional Node Examined'].quantile(0.25)  # 1. Çeyrek
Q3 = data['Regional Node Examined'].quantile(0.75)  # 3. Çeyrek
IQR = Q3 - Q1  # IQR hesaplama

# Alt ve üst sınırları hesaplama
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Aykırı değerleri belirleme
outliers = data[(data['Regional Node Examined'] < lower_bound) | (data['Regional Node Examined'] > upper_bound)]

# Aykırı değerlerin sayısını yazdırma
print(f"'Regional Node Examined' icin aykırı Değerlerin Sayısı: {outliers.shape[0]}")
```

```python
# Aykırı değerleri sınırlarla değiştirme
data['Regional Node Examined'] = np.where(
    data['Regional Node Examined'] > upper_bound, upper_bound,
    np.where(data['Regional Node Examined'] < lower_bound, lower_bound, data['Regional Node Examined'])
)

# Aykırı değerlerin tekrar kontrolü
outliers_cleaned = data[(data['Regional Node Examined'] < lower_bound) | (data['Regional Node Examined'] > upper_bound)]
print(f"Aykırı Değerlerin Sayısı (Temizlendikten Sonra): {outliers_cleaned.shape[0]}")
```

```python
# Sayısal bir sütun için IQR hesaplama
Q1 = data['Reginol Node Positive'].quantile(0.25)  # 1. Çeyrek
Q3 = data['Reginol Node Positive'].quantile(0.75)  # 3. Çeyrek
IQR = Q3 - Q1  # IQR hesaplama

# Alt ve üst sınırları hesaplama
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Aykırı değerleri belirleme
outliers = data[(data['Reginol Node Positive'] < lower_bound) | (data['Reginol Node Positive'] > upper_bound)]

# Aykırı değerlerin sayısını yazdırma
print(f"'Reginol Node Positive' icin aykırı Değerlerin Sayısı: {outliers.shape[0]}")
```

```python
# Aykırı değerleri sınırlarla değiştirme
data['Reginol Node Positive'] = np.where(
    data['Reginol Node Positive'] > upper_bound, upper_bound,
    np.where(data['Reginol Node Positive'] < lower_bound, lower_bound, data['Reginol Node Positive'])
)

# Aykırı değerlerin tekrar kontrolü
outliers_cleaned = data[(data['Reginol Node Positive'] < lower_bound) | (data['Reginol Node Positive'] > upper_bound)]
print(f"Aykırı Değerlerin Sayısı (Temizlendikten Sonra): {outliers_cleaned.shape[0]}")
```

```python
# Sayısal bir sütun için IQR hesaplama
Q1 = data['Survival Months'].quantile(0.25)  # 1. Çeyrek
Q3 = data['Survival Months'].quantile(0.75)  # 3. Çeyrek
IQR = Q3 - Q1  # IQR hesaplama

# Alt ve üst sınırları hesaplama
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Aykırı değerleri belirleme
outliers = data[(data['Survival Months'] < lower_bound) | (data['Survival Months'] > upper_bound)]

# Aykırı değerlerin sayısını yazdırma
print(f"'Survival Months' icin aykırı Değerlerin Sayısı: {outliers.shape[0]}")
```

```python
# Aykırı değerleri sınırlarla değiştirme
data['Survival Months'] = np.where(
    data['Survival Months'] > upper_bound, upper_bound,
    np.where(data['Survival Months'] < lower_bound, lower_bound, data['Survival Months'])
)

# Aykırı değerlerin tekrar kontrolü
outliers_cleaned = data[(data['Survival Months'] < lower_bound) | (data['Survival Months'] > upper_bound)]
print(f"Aykırı Değerlerin Sayısı (Temizlendikten Sonra): {outliers_cleaned.shape[0]}")
```

```python
# Tum aykiri verileri ortak olan satir var ise bunlari kaldirmak istedik
# Fakat ortak satir 0 oldugu icin yapamadik

columns = ['Tumor Size', 'Regional Node Examined', 'Reginol Node Positive', 'Survival Months']

# Aykırı değerlerin belirlenmesi
def detect_outliers(column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (data[column] < lower_bound) | (data[column] > upper_bound)

# Her sütunda aykırı değerleri belirlenmesi
outliers_mask = pd.DataFrame(False, index=data.index, columns=columns)

for col in columns:
    outliers_mask[col] = detect_outliers(col)

# Ortak aykırı değerlerin bulunduğu satırları bulma
common_outliers = outliers_mask.all(axis=1)

# Ortak aykırı değerleri içeren satırları silme
#data_cleaned = data[~common_outliers]

# Sonuçları kontrol etme
print(f"Ortak aykırı değer bulunan satır sayısı: {common_outliers.sum()}")
#print(data_cleaned)
```

# MODELLEME

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix

# Veriyi ayirma
X = data.drop(['Status'],axis=1)
y = data['Status'].astype(int)

# Veri kümesini eğitim ve test olarak ayırma
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.17, random_state=42)

# Veriyi ölçeklendirme
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Modeli başlatma ve eğitim
model_knn = KNeighborsClassifier(n_neighbors=7)
model_knn.fit(X_train_scaled, y_train)

# Tahmin yapma ve doğruluk hesaplama
y_pred_knn = model_knn.predict(X_test_scaled)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print('Dogruluk orani:', accuracy_knn)

# Sınıflandırma raporu oluşturma
siniflandirmaRaporu = classification_report(y_test, y_pred_knn)
print('siniflandirma Raporu:', siniflandirmaRaporu)

# Confusion Matrix oluşturma
cm = confusion_matrix(y_test, y_pred_knn)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Negatif', 'Pozitif'], yticklabels=['Negatif', 'Pozitif'])
plt.xlabel('Tahmin Edilen Değerler')
plt.ylabel('Gerçek Değerler')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

# KNN modelinin pozitif sınıf olasılıklarını alıyoruz
y_pred_proba_knn = model_knn.predict_proba(X_test_scaled)[:, 1]  # Pozitif sınıf olasılıkları

# ROC eğrisini çizmek için fpr (false positive rate) ve tpr (true positive rate) hesaplama
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba_knn)

# AUC değerini hesaplama
auc = roc_auc_score(y_test, y_pred_proba_knn)

# ROC eğrisini çizme
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"KNN (AUC = {auc:.2f})")
plt.plot([0, 1], [0, 1], 'k--')  # Rastgele tahmin çizgisi
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - KNN")
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

```python
import tensorflow as tf

# Veriyi hazırlıyoruz
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

# Veriyi eğitim ve test olarak ayırıyoruz
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.17, random_state=42)

# Veriyi ölçeklendiriyoruz
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Yapay Sinir Ağı modelini oluşturuyoruz
ann = tf.keras.models.Sequential()

# İlk katman olarak Input kullanıyoruz
ann.add(tf.keras.layers.Input(shape=(X_train_scaled.shape[1],)))  # İlk katman

# Gizli katmanlar ve dropout katmanları
ann.add(tf.keras.layers.Dense(units=64, activation="relu"))
ann.add(tf.keras.layers.Dropout(0.2))  # Dropout ile aşırı öğrenmeyi engelliyoruz
ann.add(tf.keras.layers.Dense(units=32, activation="relu"))
ann.add(tf.keras.layers.Dropout(0.2))  # Dropout
ann.add(tf.keras.layers.Dense(units=16, activation="relu"))

# Çıkış katmanı
ann.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Modeli derliyoruz
ann.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

# Erken durdurma (Early stopping) ekliyoruz
early_stopping = tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True)

# Modeli eğitiyoruz
history = ann.fit(X_train_scaled, y_train, epochs=100, validation_split=0.2, batch_size=32, callbacks=[early_stopping])

# Test verisi ile tahmin yapıyoruz
y_pred = ann.predict(X_test_scaled)
y_pred = (y_pred > 0.5)

# Modelin doğruluğunu ve confusion matrix'i hesaplıyoruz
accuracy_ann = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Test doğruluğu: {accuracy_ann * 100:.2f}%")
print(f"Confusion Matrix:\n{conf_matrix}")
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np

# Veriyi hazırlıyoruz
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

# Veriyi eğitim ve test olarak ayırıyoruz
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.17, random_state=42)

# Veriyi ölçeklendiriyoruz
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# LSTM girişi için veriyi yeniden şekillendiriyoruz
X_train_scaled = X_train_scaled.reshape((X_train_scaled.shape[0], X_train_scaled.shape[1], 1))
X_test_scaled = X_test_scaled.reshape((X_test_scaled.shape[0], X_test_scaled.shape[1], 1))

# LSTM modelini oluşturuyoruz
lstm_model = tf.keras.models.Sequential()

# Giriş katmanını belirtiyoruz
lstm_model.add(tf.keras.layers.Input(shape=(X_train_scaled.shape[1], 1)))

# İlk LSTM katmanı ve dropout
lstm_model.add(tf.keras.layers.LSTM(units=64, activation="tanh", return_sequences=True))
lstm_model.add(tf.keras.layers.Dropout(0.2))

# İkinci LSTM katmanı ve dropout
lstm_model.add(tf.keras.layers.LSTM(units=32, activation="tanh", return_sequences=False))
lstm_model.add(tf.keras.layers.Dropout(0.2))

# Çıkış katmanı
lstm_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Modeli derliyoruz
lstm_model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

# Erken durdurma (Early stopping) ekliyoruz
early_stopping = tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True)

# Modeli eğitiyoruz
history = lstm_model.fit(X_train_scaled, y_train, epochs=100, validation_split=0.2, batch_size=32, callbacks=[early_stopping])

# Test verisi ile tahmin yapıyoruz
y_pred = lstm_model.predict(X_test_scaled)
y_pred = (y_pred > 0.5)

# Modelin doğruluğunu ve confusion matrix'i hesaplıyoruz
accuracy_lstm = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Test doğruluğu: {accuracy_lstm * 100:.2f}%")
print(f"Confusion Matrix:\n{conf_matrix}")
```

# AUTOENCODER

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Veriyi hazırlıyoruz
X = data.iloc[:, :-1].values  # Bağımsız değişkenler

# Veriyi eğitim ve test olarak ayırıyoruz
X_train, X_test = train_test_split(X, test_size=0.17, random_state=42)

# Veriyi ölçeklendiriyoruz
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Autoencoder için veriyi hazırlıyoruz
input_dim = X_train_scaled.shape[1]  # Giriş boyutu
encoding_dim = 14  # Gizli katman boyutu (sıkıştırma)

# Autoencoder modelini oluşturuyoruz
autoencoder = tf.keras.models.Sequential()

# Input katmanını ekliyoruz
autoencoder.add(tf.keras.layers.Input(shape=(input_dim,)))

# Encoder (sıkıştırma) katmanı ekliyoruz
autoencoder.add(tf.keras.layers.Dense(encoding_dim, activation="relu"))

# Decoder (yeniden yapılandırma) katmanı ekliyoruz
autoencoder.add(tf.keras.layers.Dense(input_dim, activation="sigmoid"))

# Modeli derliyoruz
autoencoder.compile(optimizer="adam", loss="mean_squared_error")

# Erken durdurma (EarlyStopping) ekliyoruz
early_stopping = tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True)

# Modeli eğitiyoruz ve loss metriklerini kaydediyoruz
history = autoencoder.fit(X_train_scaled, X_train_scaled, epochs=100, batch_size=2, shuffle=True,
                          validation_data=(X_test_scaled, X_test_scaled), callbacks=[early_stopping])

# Test verisi ile tahmin yapıyoruz
X_test_pred = autoencoder.predict(X_test_scaled)

# Sonuçları karşılaştırıyoruz (modelin yeniden oluşturduğu veriyi orijinal veri ile karşılaştırma)
mse = ((X_test_scaled - X_test_pred) ** 2).mean(axis=1)  # Ortalama kare hatayı hesaplıyoruz
print(f"Test Datasındaki MSE: {mse.mean()}")

# Gerçek etiketleri ve tahminleri karşılaştırmak için bir eşik belirliyoruz
threshold = mse.mean() + mse.std()  # Ortalama + Standart Sapma gibi bir eşik seçilebilir

# Tahmin edilen etiketler (anomali ise 1, değilse 0)
y_pred = (mse > threshold).astype(int)

# Gerçek etiketler (örneğin, normal=0, anomali=1)
# Bu etiketleri kendi verinizde tanımlamanız gerekir
y_true = np.random.randint(0, 2, size=len(mse))  # Burada kendi etiketlerinizi kullanın

# Accuracy hesaplıyoruz
accuracy = accuracy_score(y_true, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")

# confusion matrix 
conf_matrix = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

```
